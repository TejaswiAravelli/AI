```python
import numpy as n
import math as m
x=n.random.uniform(-10,10)
eta=0.01
max=100
i=0
while i<max:
  x_slope=4*m.pow(x,3)+6*x
  x_step_length=-eta*x_slope
  x=x+x_step_length
  i=i+1
min_value=m.pow(x,4)+3*m.pow(x,2)+10
print("control variable x =",x)
print("object function = ",min_value)
```
