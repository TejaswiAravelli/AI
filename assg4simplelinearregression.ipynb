```python

import numpy as np
import math as mt
import random
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data1.csv')
data.head()
```

```python
outliers = []
def detect_outliers_zscore(data):
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers # Driver code
```

```python
sample_outliers_X = detect_outliers_zscore(data['X'])
sample_outliers_Y = detect_outliers_zscore(data['Y'])
print("Outliers in X column from Z-scores method : ", sample_outliers_X)
print("Outliers in Y column from Z-scores method : ", sample_outliers_Y)
```

```python
print(data.isnull().sum())
```

```python
m1=data['X'].mean()
m2=data['Y'].mean()
data['X'].fillna(value=m1, inplace=True)
data['Y'].fillna(value=m2, inplace=True)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.X, data.Y, test_size = 0.10, random_state = 10)
```

```python
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.01
max_iter=10000
```

```python
def gradient(m,c,x,y):
  m_deriv=-1*(y-m*x-c)*x
  c_deriv=-1*(y-m*x-c)
  return m_deriv,c_deriv
```

```python
for iter in range(max_iter):
  for sample in range(x_train.shape[0]):
    del_m,del_c=gradient(m, c, x_train.iloc[sample], y_train.iloc[sample])
    m=m-eta*del_m
    c=c-eta*del_c
print('Model parameters m, c : ',m,c)
```

```python
y_pred_train = []
for i in x_train:
  y_p_tr = (m * i) + c
  y_pred_train.append(y_p_tr)
y_pred_train = np.array(y_pred_train)

```

```python
y_pred_test = []
for i in x_test:
  y_p_te = (m * i) + c
  y_pred_test.append(y_p_te)
y_pred_test = np.array(y_pred_test)
```

```python
import math
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error

```

```python
#Training Accuracies
mse = mean_squared_error(y_train, y_pred_train) 
print('Mean square error ', mse)
```

```python
rmse = math.sqrt(mean_squared_error(y_train, y_pred_train)) 
print('Root mean square error ', rmse)  
mae = mean_absolute_error(y_train, y_pred_train)
print('Mean ahsolute error ', mae)
```

```python
#Testing Accuracies
mse = mean_squared_error(y_test, y_pred_test) 
print('Mean square error ', mse)
rmse = math.sqrt(mean_squared_error(y_test, y_pred_test)) 
print('Root mean square error ', rmse)  
mae = mean_absolute_error(y_test, y_pred_test)
print('Mean absolute error ', mae)
```
